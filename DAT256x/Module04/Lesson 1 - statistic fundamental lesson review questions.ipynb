```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame([180.5,178.8,176.2,177.1,178.75,179.25,175.4,178.2,177.25,
                       177.5,179.9,180.8,177.75,179.5,179.25,178.5,178.8,178.5,176.5,178.76,181.6
                  ])

df.plot.hist(title='Salary Distribution', color='lightblue', bins=25)  
plt.axvline(df.mean(), color='magenta', linestyle='dashed', linewidth=2)
plt.axvline(df.median(), color='green', linestyle='dashed', linewidth=2)
plt.show()
```

A distribution is normal if the frequencies of the values form a symmetric bell-shaped curve. It is right-skewed if the mean is pulled to the right, forming a tail to the right of the distribution; and it is left-skewed if the tail pulls the mean to the left. A uniform distribution has a more or less equal number of occurrences for each value.
